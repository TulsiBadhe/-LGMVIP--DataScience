# Image to Pencil Sketch with Python

# Beginner Level Task4 

```python
# import a cv2 library
import cv2

# import matplotlib 
import matplotlib.pyplot as plt
```

### Upload the Image

```python
img= cv2.imread('D:\\Sanku\\tanjiro.jpg')
```

### Display the Original Image

```python
# displaying an Actual image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Actual Image')
plt.show()
```

### Convert the image into Gray Scale

```python
gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray)
plt.title('Gray image')
plt.show()
```

### Change into Inverted Gray Image

```python
inverted=255-gray
```

```python
# displaying a inverted image
plt.imshow(inverted)
plt.title('Inverted Image')
plt.show()
```

### Blur the Image

```python
blur= cv2.GaussianBlur(inverted,(21,21),0)
```

```python
plt.imshow(blur)
plt.title('Blurred Image')
plt.show()
```

### Convert the Image into Pencil Sketch

```python
PS=cv2.divide(gray,255-blur,scale=256)
```

```python
# to download the Pencil Sketch Image
cv2.imwrite('pencilimg.jpg',PS)
```

```python
# displaying a pencil sketch image
plt.imshow(PS,cmap="gray")
plt.title('Pencil Sketch Image')
plt.show()
```

### Displaying the Original and Pencil Sketch Image

```python
plt.imshow(PS,cmap="gray")
plt.title('Pencil Sketch Image')
plt.show()
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.show()
```
